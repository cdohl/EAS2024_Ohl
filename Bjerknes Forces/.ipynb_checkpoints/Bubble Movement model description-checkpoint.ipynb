## Equations of motion for large driving

We use the Keller-Miksis model, Eq. (6) for the volume oscillation of the bubble. It accounts for energy dissipation through viscousity and acoustic radiation losses. Because of the stronger driving bubbles radiate sound leading to coupling of neighbouring bubbles $(i,j)$ through the sound emission. This interaction is approximated with the last term in Eq. (6). The second last term is the Bernoulli pressure generated by the translation of the bubble.

$$
\left(1-\frac{\dot{R_i}}{c}\right)R_i \ddot{R}_i+
\left( \frac{3}{2}-\frac{\dot{R_i}}{2c} \right) R_i^2=\\
\frac{1}{\rho}\left(1-\frac{\dot{R_i}}{c}\right)\left[p(R_i)-p_l(t)\right]+
\frac{R_2}{\rho c}\fud{}{t}\left[p(R_i)-p_l(t)\right]\\+
\frac{1}{4}\vb{u}_i \cdot \vb{u}_i-
\sum_{i\ne j} \frac{1}{r_{ij}}\left( R_j^2\ddot{R_j}+2 R_j \dot{R_j}^2\right)
\tag{6}$$

The pressure in the liquid just outside the gas bubble $p(R_i)$ is related with the gas pressure by the jump condition due to normal viscous stresses and the Laplace pressure:

$$
p_l(R_i)=\left( p_\infty+\frac{2\sigma}{R_i^{eq}}\right)
\left( \frac{R_i^{eq}}{R_i}\right)^{3\gamma}-
\frac{2\sigma}{R_i}-\frac{4\mu\dot{R}_i}{R_i}\tag{7}
$$

The pressure in the liquid far away from the bubble contains the static pressure $p_\infty$ and the time dependent acoustic pressure $p_l(t)=p_\infty + p(t)$.

Besides the radial dynamics we also want to model their translation. Mettin et al. [] have developed and tested a simple approach that abstracts the fluid mechanics into individual forces acting on the bubbles an its displaced fluid mass. This approach can be written as Newton's 2nd law for each bubble $i$

$$
\fud{}{t}\left(M_i(t) \va{U}_i(t)\right) = \fud{}{t}\left( \frac{1}{2} \rho V_i(t) \va{U}_i(t)\right) = \va{F}_{B,i} + \va{F}_{D,i}\quad, \tag{8}
$$

stating the change of momentum of the bubble's momentum, L.H.S. of Eq. (8), is caused by the forces acting on the bubble, R.H.S of Eq. (8). For the present problem we only consider secondary Bjerkes force, $F_{B,i}$, and the drag force $F_{D,i}$. Note that the mass of the bubble content is ignored and only the added mass of the liquid is accounted for.

We can rewrite Eq. (8) into a differential equation for the position of the bubble $\va{x}$

$$\ffud{\va{x}_i}{t}=\fud{\va{U}_i}{t}=\frac{1}{R_i^3}\left[ \frac{3}{2}\frac{1}{\pi \rho}
\left(\va{F}_{B,i}+\va{F}_{D,i}\right)-3R^2\dot{R} \va{U}_i\right]\tag{9}
$$

By integrating Eq. (9) together with Eq. (6) we obtain a full set of equations to solve for the bubble position as a function of time $\va{x}_i(t)$.

We use a formulation for the drag force $\va{F}_{D,i}$ from Mettin et al. 

$$\vb{F}_{D,i}=-12\pi\mu R_i \vb{U}\tag{10}.$$

The last important part is to deal with bubbles coming into contact. In experiments we see that bubbles attracting each other will remain for some time in close contact and may coalesce. Force close distances the interaction term in Eq. (6) needs correction, typically by adding corrections of higher order in the underlying potential flow model [cite a few]. Yet for bubbles in close contact are separated by a thin liquid film undergoing oscillatory flow. A proper description would need to take into account viscous effects. Here, we ignore the details of bubbles oscillating at close distance and to prevent coalesence the Bjerknes force becomes repulsive once the bubbles are closer than $R^{eq}_i+R^{eq}_j$. This is achieved by multiplying a factor $\alpha$ to the Bjerknes force $\va{F}_{B,i}$

$$
\alpha=\tanh\left[100\left(\frac{d_{ij}}{R^{eq}_i+R^{eq}_j}-1\right) \right]\tag{11}
$$

## Simulation Results

We start with a study the magnitude and direction of the Bjerknes force for bubbles driven at $f=1.1\,$MHz and in the size range from $1\le R^{eq}\le10\,\mu$m. For this we calculate the Bjerknesforce coefficient, $f_B$, averaged over one oscillation cycle:

$$
f_B=\frac{\rho}{4\pi}\left<\dot{V}_1(t) \dot{V}_2(t)\right>\quad .
\tag{12}$$

Here we use the water as a liquid with $\mu=0.001\,$Pa$\cdot$s, $\rho=998\,$kg$\,$m$^{-3}$, $\gamma=1.4$, and $\sigma=0.072\,$N$\,$m$^{-1}$. A positive coefficient is an attractive force, while a negative coefficient is repulsive. The coefficient varies over many orders of magnitude. Therefore, we plot not the value but the logarithm of the coefficient $\log_{10}|f_B|$ in Fig. M1. Positive coefficients are colored in red and negative (repulsive bubbles) in blue. The three frames of Fig. M1 cover the radii parameter space for the acoustic pressures $p_a$ between $0.1\,$bar and $1.9\,$bar. Values very close to zero $\log_{10}|f_B|<-21$ are omitted. The linear resonance radius is indicated with a circle in Fig. M1a for guidance. Although $f_B$ does not explicitely depend on the distance between the bubbles $d_{12}$ the radial dynamics does. As we are interested in bubbles oscillating nearby each other we fixed the distance to $d_{12}=2(R_1^{eq}+R_2^{eq})$ for the calculations shown in Fig. M1.

For low driving $p_a=0.1\,$bar two distinct rectangular regions show attraction and repulsion, while for larger pressures these regions develop as a result of the nonlinearities of the bubble oscillations convoluted boundaries. Although the details of the boundaries change the general regions of mutual attraction and repulsion do not.

An example to discuss the motion of an ensemble of bubbles due to the Bjerknes force is shown in Fig. M2. Five of them have a radius of $R_1^{eq}=1.5\,\mu$m and one bigger has a radius of $R_2^{eq}=5\,\mu$m; thus their radii are below and above the linear resonance size. We expect attraction between the small bubbles and repulsion of the small bubbles away from the large bubble (acoustic pressue is $0.1\,$bar). The simulationn covers 6,000 cycles. Figure M2b depicts the trajectory of all 6 bubbles. The bigger bubble moved to the right while the smaller move first away from the big bubble but later as the repulsive force reduces form a cluster of small bubbles due to their attractive force. A rough estimate of the averaged speed of the bubbles can be obtained from the balance of the drag force, $\va{F_D}$ with the secondary Bjerknes force $\va{F_B}$.

$$
U_{avg}=\frac{f_b}{12 \,d_{12}^2 \pi \mu R}\tag{}
$$

The velocity of $R_2$ can be estimated by picking $f_b=5\times 3\cdot10^{-20}\,$kg$\,$m$^3\,$s$^{-2}$ from Fig. M2a and using an averaged distance of the big bubble to the 5 smaller bubbles of $d_{12}=15\,\mu$m. Inserting these values into Eq. (13) we obtain a velocity of $U_{avg}=3.5\,$mm/s or a distance of $19\mu$m travelled over the 6,000 cycles, which agrees approximately with the distance seen in Fig. M2b.
